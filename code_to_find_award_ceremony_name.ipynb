```python
import json
import re
def find_ceremony_name(tweet_data):
  hashtag_dict = {}
  for tweet in tweet_data:
    text = tweet['text']
    hashtags = re.findall('#([a-zA-Z0-9_]{1,50})',text)
    for htag in hashtags:
      htag = htag.lower()
      if htag in hashtag_dict.keys():
        hashtag_dict[htag] += 1
      else:
        hashtag_dict[htag] = 1
  hashtag_dict = sorted(hashtag_dict.items(), key = lambda k: -k[1])
  hashtag_dict = dict(hashtag_dict)
  return hashtag_dict
tweet_data = open_jsonfile()
find_ceremony_name(tweet_data)
```
